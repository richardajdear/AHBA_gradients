```python
from itertools import combinations

def get_region_ds(expression_donors):
    region_corrs = {}
    
    for x,y in combinations(range(6), 2):
        X = expression_donors[x]
        Y = expression_donors[y]

        name=f'{x}{y}'
        region_corrs[name] = X.T.corrwith(Y.T)
    
    region_ds = pd.concat(region_corrs, axis=1).mean(axis=1)[:180]
    
    return region_ds
```

```python
region_ds_dict = {
    '2+ DS9': get_region_ds(expression_donors_ds9),
    # '3+ DS5': get_region_ds(expression_donors_ds5)
}

region_ds = (
    pd.concat(region_ds_dict, axis=1)
    .apply(lambda x: (x-np.mean(x))/np.std(x))
    .join(get_labels_hcp())
)

region_ds_plot = region_ds.melt(id_vars='label', var_name='version', value_name='Region DS').dropna()
```

```python
def make_scatter(version, region_ds):
    out = (
        version.clean_scores().set_index('label')
        .join(region_ds.set_index('label'))
        .reset_index()
        .melt(id_vars=['label','G1','G2','G3'], value_name='region_ds', var_name='region_ds_version')
        .melt(id_vars=['label','region_ds_version','region_ds'], value_name='G_score', var_name='G')
        # .loc[:,['G2', region_ds_version]]
        # .rename({region_ds_version:'region_ds'},axis=1)
    )
    return out

# region_ds_scatter = {
#     # 'DS9': make_scatter(dm_hcp_ds9, '2+ DS9'),
#     '3+ DS5': make_scatter(dm_hcp_3d_ds5, '3+ DS5')
# }

# region_ds_scatter = (
#     pd.concat(region_ds_scatter)
#     .reset_index(0).rename({'level_0':'version'},axis=1)
# )
region_ds_scatter = make_scatter(dm_hcp_3d_ds5, region_ds.drop('2+ DS9', axis=1))
region_ds_scatter
```

```python
%%R -w 5 -h 10 --units in -r 100 -i region_ds_scatter
corrs <- region_ds_scatter %>%
    group_by(region_ds_version, G) %>% summarize(cor(G_score, region_ds, use='p')) %>%
    mutate(sig = c('', '', '***'))
    # mutate(version = factor(version, ordered=T, levels=unique(.$version)))
names(corrs) <- c('region_ds_version', 'G', 'r', 'sig')

region_ds_scatter %>%
# mutate(version = factor(version, ordered=T, levels=unique(.$version))) %>%
ggplot(aes(x=region_ds, y=G_score)) +
facet_grid(G~region_ds_version, switch='y') +
geom_point(alpha=.5) +
geom_smooth(color='mediumturquoise', method='lm') +
geom_text(data=corrs, x=-1, y=2.8, aes(label=paste('r =', round(r, 2), sig)), size=6) +
ylab('') + xlab('Mean inter-donor variation') +
theme_minimal() +
theme(aspect.ratio=1, 
      strip.placement='outside',
      strip.text.y.left=element_text(angle=0),
      panel.grid.minor = element_blank(),
      text=element_text(size=20))#, panel.spacing=unit(18, 'lines'))
```
